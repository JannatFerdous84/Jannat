```python
# Import necessary libraries
from sklearn.datasets import load_breast_cancer  # Load dataset
from sklearn.model_selection import train_test_split  # Splitting data
import pandas as pd  # For dataframes
import numpy as np  # For numerical operations
```

```python
SEED = 42
```

```python
# Load the breast cancer dataset
cancer_data = load_breast_cancer()
```

```python

# Create a DataFrame with feature names
df = pd.DataFrame(cancer_data.data, columns=cancer_data.feature_names)
```

```python
# Add the target variable to the DataFrame
df['target'] = cancer_data.target
```

```python
# Split the data into training (60%), validation (20%), and test (20%) sets
train_df, temp_df = train_test_split(df, test_size=0.2, random_state=SEED)
val_df, test_df = train_test_split(temp_df, test_size=0.5, random_state=SEED)
```

```python
# Verify dataset sizes
print("Training set size:", train_df.shape)
print("Validation set size:", val_df.shape)
print("Test set size:", test_df.shape)
```

```python
def classify_silly(sample):
    if sample.name % 2 == 0:  # If index is even, classify as malignant
        return 1
    else:  # If index is odd, classify as benign
        return 0
```

```python
# Apply the rule to the validation dataset
val_df['predicted'] = val_df.apply(classify_silly, axis=1)
```

```python
# Check accuracy
val_accuracy = (val_df['predicted'] == val_df['target']).mean()
print(f"Validation Accuracy (Silly Rule): {val_accuracy:.2f}")
```

```python
print(train_df[['mean radius', 'target']].groupby('target').mean())
```

```python
def classify_sensible(sample):
    if sample['mean radius'] > 14:
        return 1  # Malignant
    else:
        return 0  # Benign
```

```python
# Apply the rule to the validation dataset
val_df['predicted'] = val_df.apply(classify_sensible, axis=1)
```

```python
# Check accuracy
val_accuracy = (val_df['predicted'] == val_df['target']).mean()
print(f"Validation Accuracy (Sensible Rule): {val_accuracy:.2f}")
```

 If worst compactness > 0.3, the target should be 0.

```python
# Define classification rule
def classify_rule1(sample):
    if sample['worst compactness'] > 0.3:
        return 0  # Predict benign (0)
    else:
        return 1  # Predict malignant (1)
```

```python
# Apply the rule to the validation dataset
val_df['predicted'] = val_df.apply(classify_rule1, axis=1)
```

```python
# Check accuracy
val_accuracy = (val_df['predicted'] == val_df['target']).mean()
print(f"Validation Accuracy (Rule 1): {val_accuracy:.2f}")
```

```python
# Apply rule to test set
test_df['predicted'] = test_df.apply(classify_rule1, axis=1)
```

```python
# Check accuracy on test set
test_accuracy = (test_df['predicted'] == test_df['target']).mean()
print(f"Final Test Accuracy: {test_accuracy:.2f}")
```

```python

with open("results.txt", "w") as f:
    f.write(f"Validation Accuracy (Rule 1): {val_accuracy:.2f}\n")
    f.write(f"Final Test Accuracy: {test_accuracy:.2f}\n")
print("Results saved to results.txt")
```
